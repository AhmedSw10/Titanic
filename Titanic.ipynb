
🔍 ***Insights Analysis ***

**Age Distribution**

The majority of passengers were young adults, though a considerable number of children and elderly individuals were also present. This diverse age distribution influenced survival rates, especially among the most vulnerable age groups.


**Gender and Survival**

Women exhibited significantly higher survival rates compared to men, aligning with the maritime principle of "women and children first." This policy had a clear impact on the overall demographics of survivors.

**Passenger Class and Survival**

Survival rates were notably higher among first-class passengers than those in second or third class. This highlights a clear class-based disparity, likely influenced by cabin location, access to lifeboats, and prioritization during evacuation.

**Family Size and Survival**

Passengers traveling with small to moderate family sizes had better chances of survival, possibly due to increased mutual support. In contrast, individuals traveling alone had the lowest survival rates, suggesting isolation reduced their chances of receiving timely assistance.

**Embarkation Town and Survival**
Survival rates varied depending on the town of embarkation, potentially reflecting differing socioeconomic statuses and passenger profiles. Notably, Cherbourg stood out as the embarkation point with the highest number of survivors. As a coastal city, Cherbourg may have attracted wealthier passengers who secured first-class accommodations, thereby increasing their likelihood of survival.
